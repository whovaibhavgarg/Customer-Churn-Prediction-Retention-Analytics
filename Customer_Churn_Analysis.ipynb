Feature Based Analysis


```python
df['tenure_bucket']=pd.cut(
                    df['Tenure Months'],
                    bins=[0,6,12,24,60],
                    labels=['0-6','6-12','12-24','24+']
)
```

```python
df['tenure_bucket'].value_counts().sort_index()
```

```python
df['early_tenure'] = (df['Tenure Months'] <= 6).astype(int)
```

```python
df['is_month_to_month'] = (df['Contract'] == 'Month-to-month').astype(int)
```

```python
df['manual_payment'] = df['Payment Method'].isin(
    ['Electronic check', 'Mailed check']
).astype(int)
```

```python
df['Total Charges'] = pd.to_numeric(
    df['Total Charges'],
    errors='coerce'
)
```

```python
df['avg_monthly_revenue'] = df['Total Charges'] / df['Tenure Months']
```

```python
df['avg_monthly_revenue'] = (
    df['avg_monthly_revenue']
        .replace([float('inf'), -float('inf')], 0)
        .fillna(0)
)
```

```python
df[['Total Charges', 'Tenure Months', 'avg_monthly_revenue']].head()
```

```python
city_churn_rate = (
    df.groupby('City')['Churn']
      .mean()
)

df['city_churn_rate'] = df['City'].map(city_churn_rate)
city_churn_df = (
    df
    .groupby('City', as_index=False)
    .agg(
        city_churn_rate=('Churn', 'mean'),
        total_customers=('Churn', 'count')
    ).sort_values('city_churn_rate',ascending=False)
)

city_churn_df.head()

```

```python
protection_cols = [
    'Online Security',
    'Online Backup',
    'Device Protection',
    'Tech Support'
]

df['protection_score'] = (
    df[protection_cols] == 'Yes'
).sum(axis=1)
```

```python
df['streaming_user'] = (
    (df['Streaming TV'] == 'Yes') |
    (df['Streaming Movies'] == 'Yes')
).astype(int)

```

```python
df[[
    'Tenure Months', 'tenure_bucket', 'early_tenure',
    'is_month_to_month', 'manual_payment',
    'avg_monthly_revenue', 'city_churn_rate',
    'protection_score', 'streaming_user',
    'Churn'
]].head()
```

Model Prep and Encoding

```python
y = df['Churn']
X = df.drop(columns=['Churn', 'City'], errors='ignore')
```

```python
categorical_cols = [
    'Gender',
    'Senior Citizen',
    'Partner',
    'Dependents',
    'Phone Service',
    'Multiple Lines',
    'Internet Service',
    'Online Security',
    'Online Backup',
    'Device Protection',
    'Tech Support',
    'Streaming TV',
    'Streaming Movies',
    'Contract',
    'Paperless Billing',
    'Payment Method',
    'tenure_bucket'
]
```

```python
X_encoded = pd.get_dummies(
    X,
    columns=categorical_cols,
    drop_first=True
)
```

```python
X_encoded.shape
X_encoded.isnull().sum().sum()
```

```python
X_encoded.isnull().sum().sort_values(ascending=False).head(10)
```

```python
df['Total Charges'] = df['Total Charges'].fillna(0)
```

```python
X = df.drop(columns=['Churn', 'City'], errors='ignore')

X_encoded = pd.get_dummies(
    X,
    columns=categorical_cols,
    drop_first=True
)
```

```python
X_encoded.isnull().sum().sum()
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,y_train,y_test= train_test_split(
        X_encoded,
        y,
        test_size=0.2,
        random_state=42,
        stratify=y
)
    
```

```python
X_train.shape, X_test.shape
```

TESTING the Models


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_auc_score
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=42
)

log_reg.fit(X_train_scaled, y_train)
```

```python
y_pred = log_reg.predict(X_test_scaled)
y_prob = log_reg.predict_proba(X_test_scaled)[:, 1]
```

```python
log_reg.coef_
coef_df = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Coefficient': log_reg.coef_[0]
})

coef_df.sort_values(by='Coefficient', ascending=False)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
roc_auc_score(y_test, y_prob)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, precision_score, recall_score

def evaluate_threshold(threshold):
    y_pred_thresh = (y_prob >= threshold).astype(int)
    
    cm = confusion_matrix(y_test, y_pred_thresh)
    precision = precision_score(y_test, y_pred_thresh)
    recall = recall_score(y_test, y_pred_thresh)
    
    return {
        'threshold': threshold,
        'precision': precision,
        'recall': recall,
        'false_negatives': cm[1, 0],
        'false_positives': cm[0, 1],
        'customers_flagged': y_pred_thresh.sum()
    }
```

```python
thresholds = np.arange(0.3, 0.61, 0.05)

results = [evaluate_threshold(t) for t in thresholds]

threshold_df = pd.DataFrame(results)
threshold_df
```

```python
final_threshold = 0.4

y_pred_final = (y_prob >= final_threshold).astype(int)
print(confusion_matrix(y_test, y_pred_final))
print(classification_report(y_test, y_pred_final))
```

```python
#copy of original df rows corresponding to X_test
risk_df = df.loc[X_test.index].copy()
```

```python
risk_df['churn_probability'] = y_prob
risk_df['churn_prediction'] = (y_prob >= 0.4).astype(int)
```

```python
customer_risk_table = risk_df[
    [
        'City',
        'Tenure Months',
        'Contract',
        'Payment Method',
        'avg_monthly_revenue',
        'early_tenure',
        'is_month_to_month',
        'manual_payment',
        'churn_probability',
        'churn_prediction',
        'Churn'   
    ]
]
```

```python
customer_risk_table = customer_risk_table.sort_values(
    by='churn_probability',
    ascending=False
)
customer_risk_table.head(10)

```

```python
import sqlite3

conn = sqlite3.connect("Customer_Churn_Analysis.db")

customer_risk_table.reset_index(drop=True, inplace=True)
customer_risk_table['customer_id'] = customer_risk_table.index + 1

customer_risk_table.rename(columns={
    'Tenure Months': 'tenure_months',
    'Payment Method': 'payment_method',
    'Churn': 'actual_churn'
}, inplace=True)

customer_risk_table.to_sql(
    'customer_churn_risk',
    conn,
    if_exists='replace',
    index=False
)
customer_risk_table.to_csv(
    "customer_churn_risk.csv",
    index=False,
    encoding="utf-8"
)
```

```python
import sqlite3

conn = sqlite3.connect("Customer_Churn_Analysis.db")
cursor = conn.cursor()

cursor.executescript("""
DROP TABLE IF EXISTS churn_segment_summary;

CREATE TABLE churn_segment_summary AS
SELECT
    contract,
    payment_method,
    COUNT(*) AS total_customers,
    ROUND(AVG(actual_churn), 3) AS churn_rate,
    ROUND(AVG(churn_probability), 3) AS avg_churn_probability
FROM customer_churn_risk
GROUP BY contract, payment_method;
""")

conn.commit()
```
